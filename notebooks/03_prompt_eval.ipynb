### 정확도 측정

```python
import os
import sys
import duckdb

sys.path.append('..')
os.chdir('..')

from agent.agent import PayInsightAgent


def evaluate_sql(generated_sql: str, expected_sql: str) -> dict:
    """생성된 SQL과 기대 SQL의 결과를 비교"""
    con = duckdb.connect('data/ecommerce_payinsight.duckdb', read_only=True)
    try:
        gen_df = con.execute(generated_sql.strip()).df()
        exp_df = con.execute(expected_sql.strip()).df()

        if gen_df.shape[1] != exp_df.shape[1]:
            return {"pass": False, "reason": f"컬럼 수 불일치 (생성:{gen_df.shape[1]} / 기대:{exp_df.shape[1]})"}
        if gen_df.shape[0] != exp_df.shape[0]:
            return {"pass": False, "reason": f"행 수 불일치 (생성:{gen_df.shape[0]} / 기대:{exp_df.shape[0]})"}

        return {"pass": True, "reason": "OK"}

    except Exception as e:
        return {"pass": False, "reason": f"SQL 오류: {str(e)[:80]}"}
    finally:
        con.close()


def run_benchmark(agent, benchmark: list, version: str = "v1") -> dict:
    """전체 벤치마크 실행 - SQL 정확도만 측정"""
    from agent.prompt import FEW_SHOT_EXAMPLES

    results = []

    for qa in benchmark:
        # 인사이트 스킵하고 SQL만 생성 (속도 향상)
        context      = agent.retriever.retrieve(qa['question'])
        sql_response = agent.sql_chain.invoke({
            "question": qa['question'],
            "few_shot": FEW_SHOT_EXAMPLES,
            "context" : context
        })
        generated_sql = sql_response.content.strip()

        eval_result = evaluate_sql(generated_sql, qa['expected_sql'])
        results.append({
            "id"           : qa['id'],
            "level"        : qa['level'],
            "question"     : qa['question'],
            "generated_sql": generated_sql,
            "expected_sql" : qa['expected_sql'],
            "pass"         : eval_result['pass'],
            "reason"       : eval_result['reason']
        })

        status = "[O]" if eval_result['pass'] else "[X]"
        print(f"  {status} [{qa['id']}] {qa['question'][:35]}")

    total    = len(results)
    passed   = sum(1 for r in results if r['pass'])
    accuracy = passed / total * 100

    print(f"\n{'='*50}")
    print(f"[{version}] 정확도: {passed}/{total} ({accuracy:.1f}%)")
    for level in ['단순', '조건필터', '멀티분석']:
        lvl = [r for r in results if r['level'] == level]
        print(f"  {level:8s}: {sum(1 for r in lvl if r['pass'])}/{len(lvl)}")

    return {"version": version, "accuracy": accuracy, "results": results}


print("벤치마크 함수 준비 완료")
```
